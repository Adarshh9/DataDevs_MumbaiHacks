```python
def roadmap(domain ,input):
    index = domain_module_relation[domain].index(input)
    submodules_list = domain_submodule_relation[domain][index]
    print(index)
    print(input + ":")
    len_submodules = len(submodules_list)
    for i in range(len_submodules):
        print("\t" + submodules_list[i] )
        for j in range(len_submodules):
            if j==i:
                subsubmodules_list = domain_subsubmodule_relation[domain][index][j]
                for k in subsubmodules_list:
                    print("\t\t" + k)

roadmap( "Web Development" , "Frontend Development")
```

```python
def roadmap_returned(domain,input):
    data = ""
    index = domain_module_relation[domain].index(input)
    submodules_list = domain_submodule_relation[domain][index]
    data = data + (input + ":\n")
    len_submodules = len(submodules_list)
    for i in range(len_submodules):
        data = data + ("\t" + submodules_list[i] +"\n")
        for j in range(len_submodules):
            if j==i:
                subsubmodules_list = domain_subsubmodule_relation[domain][index][j]
                for k in subsubmodules_list:
                    data = data + ("\t\t" + k + "\n")
    return data
```

```python
def full_roadmap(domain,input):
    index = domain_module_relation[domain].index(input)
    len_modules = len(domain_module_relation[domain])
    while index < len_modules:
        submodules_list = domain_submodule_relation[domain][index]
        print(index)
        print(domain_module_relation[domain][index] + ":")
        len_submodules = len(submodules_list)
        for i in range(len_submodules):
            print("\t" + submodules_list[i] )
            for j in range(len_submodules):
                if j==i:
                    subsubmodules_list = domain_subsubmodule_relation[domain][index][j]
                    for k in subsubmodules_list:
                        print("\t\t" + k)
        index = index + 1

full_roadmap("Web Development" , "Frontend Development")
```

```python
def full_roadmap_returned(domain,input):
    data = ""
    index = domain_module_relation[domain].index(input)
    len_modules = len(domain_module_relation[domain])
    while index < len_modules:
        submodules_list = domain_submodule_relation[domain][index]
        data = data + (domain_module_relation[domain][index] + ":\n")
        len_submodules = len(submodules_list)
        for i in range(len_submodules):
            data = data + ("\t" + submodules_list[i] + "\n")
            for j in range(len_submodules):
                if j==i:
                    subsubmodules_list = domain_subsubmodule_relation[domain][index][j]
                    for k in subsubmodules_list:
                        data = data + ("\t\t" + k + "\n")
        index = index + 1
    return data
```
