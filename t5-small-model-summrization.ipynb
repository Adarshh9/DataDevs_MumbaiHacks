```python
import re

def preprocess_data(data):
    processed_data = []
    for sample in data:
        document = sample['document']
        summary = sample['summary']

        # Convert LaTeX expressions to plain text
        document = re.sub(r'\([^)]*\)', '', document)
        summary = re.sub(r'\([^)]*\)', '', summary)

        # Add more preprocessing steps if needed (e.g., lowercasing, removing special characters)

        processed_sample = {
            'document': document,
            'summary': summary,
            'id': sample['id']
        }

        processed_data.append(processed_sample)

    return processed_data
```

```python
data = preprocess_data(data)
data
```

```python
model_checkpoint = "t5-small"
tokenizer = AutoTokenizer.from_pretrained(model_checkpoint)
model = AutoModelForSeq2SeqLM.from_pretrained(model_checkpoint)
```

```python
# Preprocess function
max_input_length = 900
max_target_length = 180

def preprocess_function(examples):
    inputs = examples["document"]
    targets = examples["summary"]

    # Tokenize inputs and targets
    tokenized_inputs = tokenizer(inputs, max_length=max_input_length, padding="max_length", truncation=True)
    tokenized_targets = tokenizer(targets, max_length=max_target_length, padding="max_length", truncation=True)

    # Update examples with tokenized inputs and targets
    examples["input_ids"] = tokenized_inputs["input_ids"]
    examples["attention_mask"] = tokenized_inputs["attention_mask"]
    examples["labels"] = tokenized_targets["input_ids"]

    return examples
```

```python
# Tokenize and preprocess the dataset
dataset_dict = {"document": [item["document"] for item in data], "summary": [item["summary"] for item in data]}
tokenized_datasets = datasets.Dataset.from_dict(dataset_dict).map(preprocess_function, batched=True)
```

```python
# Fine-tuning configuration
batch_size = 4  # Use the entire dataset for training, as it contains only 2 examples
args = Seq2SeqTrainingArguments(
    "custom-finetuned-model",
    evaluation_strategy="epoch",
    learning_rate=1e-5,
    per_device_train_batch_size=batch_size,
    per_device_eval_batch_size=batch_size,
    save_total_limit=3,
    num_train_epochs=50,
    predict_with_generate=True,
    fp16=True,
)
```

```python
# Data collator
data_collator = DataCollatorForSeq2Seq(tokenizer, model=model)
```

```python
# Trainer
trainer = Seq2SeqTrainer(
    model,
    args,
    train_dataset=tokenized_datasets.map(lambda x: {"input_ids": x["input_ids"], "attention_mask": x["attention_mask"], "labels": x["labels"]}),
    eval_dataset=tokenized_datasets.map(lambda x: {"input_ids": x["input_ids"], "attention_mask": x["attention_mask"], "labels": x["labels"]}),
    data_collator=data_collator,
    tokenizer=tokenizer,
#     compute_metrics=compute_metrics
)
```

```python
# Fine-tune the model
trainer.train()
```
